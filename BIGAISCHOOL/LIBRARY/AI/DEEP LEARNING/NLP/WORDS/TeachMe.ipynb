```python
import os
import re
import random
import requests
from bs4 import BeautifulSoup
```

```python
files = [f.name for f in os.scandir('/home/aleksander/PycharmProjects/langbang/DATA/FLASHCARDS/GERMAN/FROM_PL/SLOWNICTWO1') if f.is_file()]
```

```python
files
```

```python
words = []
for word in files:
    oko = re.sub(".*\\s", "", word)
    last  = os.path.splitext(oko)[0]
    words.append(last)  
```

```python
words_to_learn = []
for word in range(10):
    words_to_learn.append(random.choice(words))
```

```python
words_to_learn
```

```python
sentences_to_learn = []
for word in words_to_learn:
    req = requests.get("https://context.reverso.net/translation/german-english/"+word, headers={'User-Agent': 'Mozilla/5.0'})
    soup = BeautifulSoup(req.text, 'html.parser')
    sentences = [x.text.strip() for x in soup.find_all('span', {'class':'text'}) if '\n' in x.text]
    sentences_to_learn.append(sentences)
```
